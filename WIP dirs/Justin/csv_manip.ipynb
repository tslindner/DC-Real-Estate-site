```python
import pandas as pd
import numpy as np

df = pd.read_csv("input_data/real_estate.csv")

cols = df.columns
my_dict = dict()

for i in cols:
    if f"{i}" not in my_dict.keys():
        my_dict[f"{i}"] = []
        
    try:
        if type(df[i].max()) != str and type(df[i].min()) != str:
            print(i)

            my_dict[f"{i}"].append(df[i].min())
            my_dict[f"{i}"].append(df[i].max())
        else: 

            my_dict[f"{i}"].append(0)
            my_dict[f"{i}"].append(0)
    except:
        my_dict[f"{i}"].append(0)
        my_dict[f"{i}"].append(0)
        
min_max = pd.DataFrame(my_dict).fillna(0)

final_df = pd.concat([min_max, df], ignore_index=True)
```
